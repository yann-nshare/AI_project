**Exercice 1.3.2:**
Soit les matrices $A$ et $B$ suivantes:
$$A = \begin{bmatrix} 1 & 2 & 1 \\ 3 & 0 & 1 \\ 0 & 2 & 4 \end{bmatrix}$$
$$B = \begin{bmatrix} 2 & 0 & 1 \\ 1 & 2 & 3 \\ 4 & 0 & 1 \end{bmatrix}$$
Construisez un programme en python permettant de multiplier 2 matrices.

**INTERDIT** d'utiliser une librairie externe ou l'opérateur `**` ! :)

````python

def multiply(a, b):
    result = [[0,0,0],
             [0,0,0],
             [0,0,0]]
    ### Début du code ###
    for i in range(len(a)):
        result[i] = [a[i][l] * b[i][l] for l in range(len(a[i]))]
    ### Fin du code ###
    for r in result:
       print(r)
    
a = [[1, 2, 1],
     [3, 0, 1],
     [0, 2, 4]]
b = [[2, 0, 1],
     [1, 2, 3],
     [4, 0, 1]]
multiply(a, b)
````

**Résultat attendu:**  
`[8, 4, 8]
[10, 0, 4]
[18, 4, 10]`

## 2.4. Introduction à la descente de gradient

La descente de gradient fait partie des notions de bases du Machine Learning. Elle a pour but de minimiser une fonction. En effet, un algorithme de Machine Learning est une fonction mathématique (comme ce que vous avez vu dans les exercices précédents) plus ou moins complexe. Elle possède des variables inconnues ($x$, $y$, $z$ par exemple) et votre but sera de déterminer ces variables. On appelle ces variables les poids, le biais, etc... les paramètres du modèle. 
Au début, vous initialisez vos paramètres de manière aléatoire, puis votre modèle calculera une valeur d'erreur en comparant vos prédictions et les vraies résultats. Enfin, après avoir calculé les gradients de vos paramètres, votre modèle se chargera de les updater petit à petit jusqu'à ce qu'ils obtiennent les bonnes valeurs pour lesquelles les poids permettent d'obtenir une erreur à 0, ou la plus petite possible (d'où le "minimiser une fonction").
Ce n'est pas grave si vous ne comprenez pas tout pour le moment.
